5.  What is mutability in Python?
   - Mutability in Python refers to whether an object can be changed after its creation.

- Mutable objects: Can be modified after creation (e.g., list, dict, set).
- Immutable objects: Cannot be changed after creation (e.g., int, float, str, tuple).
6.  Why are lists mutable, but tuples are immutable?
  - Lists are mutable because they are designed for dynamic operations, allowing elements to be added, removed, or changed. This is possible because lists use an array-like structure with flexible memory allocation.

  - Tuples, on the other hand, are immutable because they are meant to be hashable and used as fixed data structures. Their immutability ensures data integrity and allows them to be used as dictionary keys or set elements, which require stable hash values.

9. What is type casting in Python?
  - Type casting in Python is the process of converting one data type into another. It can be done explicitly using built-in functions like int(), float(), str(), list(), etc., or implicitly when Python automatically converts types during operations.

- Explicit Type Casting:
num_str = "10"
num_int = int(num_str)  # Converts string to integer
print(num_int)  # Output: 10

- Implicit Type Casting:
num = 5  
decimal = 2.5  
result = num + decimal  # Python converts 'num' to float automatically
print(result)  # Output: 7.5
10. What is the difference between implicit and explicit type casting?
  - Implicit Type Casting (Automatic Casting):

- It happens automatically when converting a value from a lower data type to a higher data type (e.g., from int to float).
The compiler does it without any intervention from the programmer.
-Example:
x = 5      # integer  
y = 3.2    # float  
result = x + y  # Python automatically converts x to float
  - Explicit Type Casting (Manual Casting):
- It requires the programmer to specify the type conversion.
It happens when converting a higher data type to a lower one, which might lead to loss of precision or data.
- Example:
x = 5.6   # float  
y = int(x)  # Explicitly casting float to integer .
11. What is the purpose of conditional statements in Python?
  - Conditional statements in Python are used to make decisions in a program based on whether a certain condition is true or false. They allow the program to execute specific blocks of code depending on the outcome of a condition. The most common conditional statements are:

- if: Executes a block of code if a condition is true.
- elif: Checks another condition if the previous if condition is false.
- else: Executes a block of code if all preceding conditions are false.

This helps control the flow of execution and makes the program more dynamic.
12.  How does the elif statement work?
    - The elif statement stands for "else if" and is used in conditional programming. It follows an if statement and allows you to check multiple conditions in sequence. If the if condition is False, the program checks the elif condition. If the elif condition is True, it executes the corresponding block of code. If none of the if or elif conditions are True, the else block (if present) is executed.

- Example:

x = 10
if x > 20:
    print("x is greater than 20")

elif x == 10:
    print("x is equal to 10")

else:
    print("x is less than 10")

Here, it will print x is equal to 10 because the second condition (elif x == 10) is true.
13.  What is the difference between for and while loops?
    - A for loop is used when you know in advance how many times you need to repeat a block of code. It typically works with a sequence (like a range of numbers) or a collection (like a list).

- Example:
for i in range(5):
    print(i)

     - A while loop, on the other hand, runs as long as a certain condition is true. It is used when you don’t know in advance how many times the loop will run, but you have a condition that controls it.

- Example:
i = 0
while i < 5:
    print(i)
    i += 1
14. Describe a scenario where a while loop is more suitable than a for loop?
  - A while loop is more suitable than a for loop when the number of iterations is not known in advance and depends on a condition that is checked during the execution of the loop.

- Example Scenario:
Imagine you're processing items from a queue, and you stop when the queue is empty. You don’t know how many items will be in the queue at the start, so you keep checking if the queue has more items until it’s empty.


queue = ['item1', 'item2', 'item3']

while queue:
    item = queue.pop(0)
    print(f"Processing {item}")

Here, the number of iterations depends on how many items are left in the queue, so a while loop is a natural choice.
